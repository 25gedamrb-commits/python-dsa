```python
#stack with linked list
import time
class Node:
    def __init__(self,data=None):
        self.data= data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head=None

    def __iter__(self):
        curnode =self.head
        while curnode:
            yield curnode
            curnode=curnode.next

class stack:
    def __init__(self):
        self.Linkedlist=Linkedlist()

    def str(self):
        values=[str(x.value) for x in self.Linkedlist]
        return '\n'.join(values)

    def Isempty(self):
        if self.Linkedlist.head==None:
            return True
        else:
            return False

    def push(self,data):
        node= Node(data)
        node.next=self.Linkedlist.head
        self.Linkedlist.head=node

    def pop(self):
        if self.Isempty():
            return "there is no element"
        else:
            nodeval=self.Linkedlist.head.data
            self.Linkedlist.head=self.Linkedlist.head.next
            return nodeval

    def peek(self):
        if self.Isempty():
            return "there is no element"
        else:
            nodeval=self.Linkedlist.head.data
            return nodeval
    
    def delete():
        self.Linkedlist.head=None

    
        
c = stack()
c.push(1)
c.push(2)
c.push(3)
print(c)
print("display top element")
print(c.peek())
print("pop top element")
print(c.pop())
print(c)
print("pop top element")
print(c.pop())
print("now check stack again")
print(c)
```

```python
class Graph:
    def __init__(self):
        self.adjacency_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adjacency_list:
            self.adjacency_list[vertex] = []

    def add_edge(self, vertex1, vertex2):

        # Undirected graph
        self.adjacency_list[vertex1].append(vertex2)
        self.adjacency_list[vertex2].append(vertex1)

    def print_graph(self):
        for vertex in self.adjacency_list:
            print(f"{vertex}:{self.adjacency_list[vertex]}")

    def remove_vertex(self, vertex):
        if vertex in self.adjacency_list:
            for v in self.adjacency_list:
                if vertex in self.adjacency_list[v]:
                    self.adjacency_list[v].remove(vertex)
            del self.adjacency_list[vertex]
            return True

        return False

mygraph = Graph()

# Add vertices
mygraph.add_vertex("A")
mygraph.add_vertex("B")
mygraph.add_vertex("C")
mygraph.add_vertex("D")
mygraph.add_vertex("E")

# Add edges
mygraph.add_edge("A", "B")
mygraph.add_edge("A", "C")
mygraph.add_edge("A", "D")
mygraph.add_edge("B", "E")
mygraph.add_edge("C", "D")
mygraph.add_edge("D", "E")
mygraph.print_graph()
mygraph.remove_vertex("A")
print("removed graph")

mygraph.print_graph()
```

```python
#multilevel
class college:
    def college_name(self):
        print("modern college")

class student(college):
    def student_info(self):
        print("name: prashant jha")
        print("branch: mca")
class exam(student):
    def subject(self):
        print("subject1: maths")
        print("subject1: evs")

obj = exam()
obj.college_name()
obj.student_info()
obj.subject()
```

```python
#compile time polymorphism
#python supports operator overloading
#runtime polymorphoism
class rbi:
    def home_loan(self):
        print("home loan=8%")
    def edu_loan(self):
        print("education loan= 9%")
class Sbi(rbi):
    def edu_loan(self):
        print("education loan=10%")
        super().edu_loan()
obj=Sbi()
obj.edu_loan()
obj.home_loan()
```
