```python
#isvalid
n = "({[()]})"

stack = []

pairs = {
    ')': '(',
    '}': '{',
    ']': '['
}

valid = True

for ch in n:
    if ch in "({[":
        stack.append(ch)

    elif ch in ")}]":
        if not stack or stack[-1] != pairs[ch]:
            valid = False
            break
        stack.pop()

if valid and not stack:
    print("Valid")
else:
    print("Invalid")
        
```

```python
#insertion sort
val=[5,3,8,6,2]
for i in range(1,len(val)):
    key= val[i]
    j=i-1
    while j>=0 and val[j]> key:
            val[j+1]= val[j]
            j=j-1
    val[j+1]=key
print(val)   
```
