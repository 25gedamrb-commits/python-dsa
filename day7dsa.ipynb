```python
#factorial
def factorial(num):
    if num <= 1:
        return 1
    return num * factorial(num-1)
print(factorial(4))
```

```python
def capitalizefirst(arr):
    result=[]
    if len(arr)==0:
        return result
    result.append(arr[0][0].upper() + arr[0][1:])
    return result + capitalizefirst(arr[1:])
print(capitalizefirst(['car','taco','banana']))
```

```python
def power(base,exponent):
    if exponent==0:
        return 1
    return base* power(base,exponent-1)
print(power(2,2))
print(power(2,4))
```

```python
def productofarray(arr):
    if len(arr)==0:
        return 1
    return arr[0] * productofarray(arr[1:])
print(productofarray([1,2,3]))
print(productofarray([1,2,3,10]))
```

```python
def reverse(string):
    if len(string) ==0:
        return string
    return string[len(string)-1] + reverse(string[0:len(string)-1])
print(reverse("python"))
```

```python
def recursiverange(num):
    if num<=0:
        return 1
    return num + recursiverange(num-1)
print(recursiverange(6))
```

```python
def ispalin(string):
    if len(string)==0:
        return True
    if string[0] != string[len(string)-1]:
        return False
    return ispalin(string[1:-1])
print(ispalin("tacocat"))
```

```python
def someRecursive(arr, cb):
    if len(arr)==0:
        return False
    if not(cb(arr[0])):
        return someRecursive(arr[1:],cb)
    return True
def isodd(num):
    if num%2==0:
        return False
    else:
        return True
print(someRecursive([4,6,8],isodd))
```

```python
class Tree:
    def __init__(self,data):
        self.data=data
        self.child=[]

    def __str__(self,level=0):
        ret =" "* level + str(self.data) + "\n"
        for ch in self.child:
            ret += ch.__str__(level+1)
        return ret
        
    def addchild(self,object):
        self.child.append(object)

rootnode= Tree("Drinks")
hot= Tree("hot")
cold= Tree("cold")
tea= Tree("tea")
coffee= Tree("coffee")
nonalcoholic= Tree("nonalcoholic")
alcoholic= Tree("alcoholic")

rootnode.addchild(hot)
rootnode.addchild(cold)
hot.addchild(tea)
hot.addchild(coffee)
cold.addchild(nonalcoholic)
cold.addchild(alcoholic)

print(rootnode)
```
